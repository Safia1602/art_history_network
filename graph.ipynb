```python
import networkx as nx
import json

# Nom du fichier d'entrée et de sortie
INPUT_FILE = "network1.gexf"
OUTPUT_FILE = "graph.json"

# Lire le fichier GEXF avec NetworkX
G = nx.read_gexf(INPUT_FILE)

# Construire une structure JSON compatible avec D3.js
data = {
    "nodes": [
        {"id": str(node), **G.nodes[node]} for node in G.nodes()
    ],
    "links": [
        {"source": str(u), "target": str(v), **G.edges[u, v]}
        for u, v in G.edges()
    ]
}

# Sauvegarder au format JSON
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"Fichier converti avec succès → {OUTPUT_FILE}")
```
